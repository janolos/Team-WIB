```python
import pandas as pd
import numpy as np
```

```python
WB2020 = pd.read_csv('2020-2019 Census Business Data.csv')
```

```python
WB2020.head()
```

## Rename Columns 

```python
WB2020_1 = WB2020.rename(columns={'NAICS2017':'NAICS',
                                       'SEX':'Owners_Sex',
                                       'ETH_GROUP': 'Owners_Ethnicity', 
                                       'RACE_GROUP':'Owners_Race', 
                                       'VET_GROUP':'Vet_Status', 
                                       'RCPPDEMP':'Sales', 
                                       'EMP': 'Employees', 
                                       'RCPPDEMP_S':'Sales_Std_Error', 
                                       'EMP_S':'Employees_Std_Error'})
```

## Subset Dataset

```python
WB2020_1.head()
```

```python
WB2020_2 = WB2020_1[['NAICS', 
                     'Owners_Sex', 
                     'Owners_Ethnicity', 
                     'Owners_Race', 
                     'Vet_Status', 
                     'Sales', 
                     'Employees', 
                     'Sales_Std_Error', 
                     'Employees_Std_Error']]
```

```python
WB2020_2.head()
```

## Filter Dataframe to only contain female-owned businesses:

```python
WB2020_F = WB2020_2.Owners_Sex.str.contains('002')
WB2020_2F = WB2020_2[WB2020_F]
```

```python
WB2020_2F
```

## Determine which sectors were predominantly female-owned:

```python
WB2020_2F.NAICS.value_counts()
```

### The following sectors were top 5 in 2020: (1) Professional/Scientific/Technical Services; (2) Health Care & Social Assistance; (3) Accomodation & Food Services; (4) Manufacturing; and (5) Transportation & Warehousing.
