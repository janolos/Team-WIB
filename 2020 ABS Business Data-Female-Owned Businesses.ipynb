```python
WBiz2020 = pd.read_csv('2020-2019 Census Business Data.csv')
```

## Rename Columns 

```python
WB2020_1 = WBiz2020.rename(columns={'NAICS2017':'NAICS',
                                       'SEX':'Owners_Sex',
                                       'ETH_GROUP': 'Owners_Ethnicity', 
                                       'RACE_GROUP':'Owners_Race', 
                                       'VET_GROUP':'Vet_Status', 
                                       'RCPPDEMP':'Sales', 
                                       'EMP': 'Employees', 
                                       'RCPPDEMP_S':'Sales_Std_Error', 
                                       'EMP_S':'Employees_Std_Error'})
```

## Subset Dataset

```python
WB2020_1.head()
```

```python
WB2020_2 = WB2020_1[['NAICS', 
                     'Owners_Sex', 
                     'Owners_Ethnicity', 
                     'Owners_Race', 
                     'Vet_Status', 
                     'Sales', 
                     'Employees', 
                     'Sales_Std_Error', 
                     'Employees_Std_Error']]
```

```python
WB2020_2.head()
```

## Filter Dataframe to only contain female-owned businesses:

```python
WB2020_F = WB2020_2.Owners_Sex.str.contains('002')
WB2020_2F = WB2020_2[WB2020_F]
```

```python
WB2020_2F
```

## Convert datatype to integer 

```python
WB2020_2F.info()
```

```python
WB2020_3 = WB2020_2F.apply(pd.to_numeric, errors='coerce')
```

```python
print(WB2020_3)
```

```python
WB2020_3.info()
```

## Use Chi-Squares to determine if Owners_Race depends on Sales

### Create a contingency table:

```python
from scipy import stats
```

```python
WB2020_crosstab = pd.crosstab(WB2020_3['Owners_Race'], WB2020_3['Sales'])
```

```python
stats.chi2_contingency(WB2020_crosstab)
```

### p-value > 0.05, therefore there is not a significant relationship between owners_sex and the number of sales. The values in the array are less than 5 and therefore does not meet the assumption.

## Determine if Owners_Race depends on NAICS

```python
WB2020_crosstab2 = pd.crosstab(WB2020_3['Owners_Race'], WB2020_3['NAICS'])
```

```python
stats.chi2_contingency(WB2020_crosstab2)
```

### p-value > 0.05, therefore not a significant relationship between owner's sex and NAICS codes. The values in the array are more than 5 and it does meet the assumption.
